### Exchange Description

I want the exchange to handle two kinds of orders:
* continuous buy orders in both directions
* normal buy orders in both directions

The in order to do this, a ETH/USDT exchange must be able to access the following variables
* the current amount of stream buying in each direction
* what accounts own what percent of stream buying in each direction (for redistributing returned tokens)

When a normal buy order arrives the reserves for both assets are updated following the Uniswap V2 formula

When a buy stream is opened or closed, the following are updated
* the current amount of stream buying in that direction
* the account ownership percentages of that stream

### Simplest case: 

Let's say that there is a stream buying 100,000 USDT worth of ETH per month and another buying 10 ETH worth of USDT per month.
The exchange has 10,000,000 USDT and 2,500 ETH, setting the initial price at 4000 USDT per ETH

$R_{eth} = 2500$ ETH

$R_{usd} = 10,000,000$ USDT

This stream of exchanges is best understood as the cummulative effect of many discrete 1 second exchanges.

Within a 1 second interval, one side buys a discrete amount of ETH and the other buys a different discrete amount of USDT.
It turns out that the order executed second gets the better price.

We want to execute both orders simultaneously.

We can do so computationally by breaking both buy orders into infitesimally small pieces and executing them in sequence.

While such a function is not computationally intensive on a normal computer, it is gas intensive on the blockchain. 
Therefore, we will only use this function to show the correct result and verify a closed form solution 
or an efficient approximation

We will use this formula to calculate the amount of tokens sent within a 1 second interval.

For $t=10$ seconds since the last state change via a transaction, that amount will equal the total amount sent after 
10 seconds minus the total amount sent after 9 seconds. Thus the variable amount of tokens sent back only depends on time,
matching with Superfluid's streaming requirments.

```python
def buy(r_a, r_b, d_b, fee):
    """ buys an amount using Uniswap V2 constant product exchange formula
        inputs: reserve a, reserve b, amount b sent (delta b), fee
        returns: new reserve a, new reserve b, and amount of asset a purchased """
    d_a = r_a - (r_a * r_b / (r_b + fee * d_b))
    r_a = r_a - d_a
    r_b = r_b + d_b
    
    return r_a, r_b, d_a
```

```python
def multiswap(usd, eth, r_eth, r_usd, fee, print_results=False, NITERS=100000, usd_first=True):
    """ buy one then the other in tiny increments for fair execution 
        note: this method is computationally inefficient but provides the correct answer """
    initial_price = r_usd / r_eth
    
    dusd = usd / NITERS
    deth = eth / NITERS
    returned_usd = 0
    returned_eth = 0
    
    for i in range(NITERS):
        if usd_first:
            r_eth, r_usd, reth = buy(r_eth, r_usd, dusd, fee)
            r_usd, r_eth, rusd = buy(r_usd, r_eth, deth, fee)
        else:
            r_usd, r_eth, rusd = buy(r_usd, r_eth, deth, fee)
            r_eth, r_usd, reth = buy(r_eth, r_usd, dusd, fee)
            
        returned_eth += reth
        returned_usd += rusd
    
    if print_results:
        print("initial price: ", initial_price)
        print("usd execution price: ", usd / returned_eth)
        print("eth execution price: ", returned_usd / eth)
        print("final pool price: ", r_usd / r_eth)
        
    return returned_usd, returned_eth, r_usd, r_eth
```

## Visualize the streaming curve

```python
def month_to_second(x):
    return x / 30.5 / 24 / 60 / 60
```

```python
usdt_per_month = 5000 * 500
eth_per_month = 10 # 5000
usdt_per_month / eth_per_month
```

```python
usdt_per_second = month_to_second(usdt_per_month)
eth_per_second = month_to_second(eth_per_month)

usdt_reserves = 5000000
eth_reserves = 1250
usdt_reserves / eth_reserves
fee = .99
animation_duration = 100 * 24 * 60 * 60 * 2  # seconds
seconds_per_window = 60*60*12
```

```python
ks = []
rusds = []
reths = []

for i in range(int(animation_duration / seconds_per_window)):
    # place buy order
    usd, eth, rusd, reth = multiswap(usdt_per_second * seconds_per_window * (i+1),
                                     eth_per_second * seconds_per_window * (i+1), 
                                     eth_reserves, 
                                     usdt_reserves, 
                                     fee)
    ks.append(rusd*reth)
    rusds.append(rusd)
    reths.append(reth)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,6))
sns.despine(fig, left = True, bottom = True)
```

```python
def plot_bonding_curve(k, reth, rusd, ax=ax):
    if ax is None:
        fig, ax = plt.subplots(1,1,figsize=(10,6))
        sns.despine(fig, left = True, bottom = True)
    else:
        ax.clear()
    
    x = np.linspace(rusd*.3, rusd*1.5, num=100)
    y = k / x
    
    sns.lineplot(x=x, y=y, ax=ax)
    ax.plot(rusd, reth,'ro')
    ax.set_xlabel('USD Reserves (millions)')
    ax.set_ylabel('ETH Reserves')
    ax.set_title('Curve for {} USDT and {} ETH per month'.format(usdt_per_month, eth_per_month), fontsize=14)
    # ax.set_ylim([1000, 3000])
    # ax.set_xlim([2e6, 6e6])
    textstr = "1 ETH = ${}".format(int(rusd / reth))
    ax.text(0.8, 0.9, textstr, transform=ax.transAxes, fontsize=14)
```

```python
def animate(i):
    plot_bonding_curve(ks[i], reths[i], rusds[i], ax=ax)

ani = animation.FuncAnimation(fig, animate, frames=len(reths), interval=60, repeat=True)
ani.save('onlybuying.gif')
```
